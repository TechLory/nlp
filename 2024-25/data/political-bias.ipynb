Create a MongoDb version of the Kaggle dataset [Political Bias](https://www.kaggle.com/datasets/mayobanexsantana/political-bias) 

```python
import pymongo
import pandas as pd
```

```python
data = pd.read_csv("/Users/Flint/Data/politicalbias/Political_Bias.csv")
data.head()
```

```python
records = []
for i, row in data.iterrows():
    records.append(dict(row))
```

```python
db = pymongo.MongoClient()['news']
collection = db['politicalbias']
print(len(records))
collection.insert_many(records)
```
