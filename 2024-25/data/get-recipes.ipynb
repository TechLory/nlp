```python
import pandas as pd 
import pymongo
```

```python
recipe_file = "/Users/Flint/Data/recipes/food_recipes.csv"
R = pd.read_csv(recipe_file)
```

```python
R.head(2)
```

```python
text_fields = ['recipe_title', 'description', 'instructions']
```

```python
r_data = []
for i, row in R.iterrows():
    r = dict(row)
    for tf in text_fields:
        try:
            r[tf] = r[tf].replace('\xa0', ' ')
        except AttributeError:
            pass
    if not pd.isnull(r['ingredients']):
        r['ingredients'] = r['ingredients'].lower().split("|")
        try:
            r['tags'] = r['tags'].lower().split("|")
        except AttributeError:
            pass
        r_data.append(r)
```

```python
db = pymongo.MongoClient()['cousine']
collection = db['recipes']
collection.insert_many(r_data)
```
